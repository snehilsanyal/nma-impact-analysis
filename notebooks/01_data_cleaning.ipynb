# Data Cleaning 

## Import Libraries

```python
import os
import pandas as pd
import numpy as np

RAW_DIR = "../data/raw"
CLEAN_DIR = "../data/cleaned"
os.makedirs(CLEAN_DIR, exist_ok=True)

pd.set_option("display.max_columns", None)
pd.set_option("display.max_colwidth", 120)
```

## Load Metadata

```python
programs = pd.read_csv(f"{RAW_DIR}/programs.csv")
users = pd.read_csv(f"{RAW_DIR}/users.csv")
applications = pd.read_csv(f"{RAW_DIR}/applications.csv")
pods = pd.read_csv(f"{RAW_DIR}/pods.csv")
project_groups = pd.read_csv(f"{RAW_DIR}/project_groups.csv")
ta_assignments = pd.read_csv(f"{RAW_DIR}/ta_assignments.csv")
daily_engagement = pd.read_csv(f"{RAW_DIR}/daily_engagement.csv")
assessments = pd.read_csv(f"{RAW_DIR}/assessments.csv")
surveys = pd.read_csv(f"{RAW_DIR}/surveys.csv")
dropouts = pd.read_csv(f"{RAW_DIR}/dropouts.csv")
support_requests = pd.read_csv(f"{RAW_DIR}/support_requests.csv")
ta_feedback = pd.read_csv(f"{RAW_DIR}/ta_feedback.csv")
pod_checkins = pd.read_csv(f"{RAW_DIR}/pod_checkins.csv")
```

## Audit Dataframes

```python
def audit_df(df, name):
    print(f"\n{'='*80}")
    print(f"{name.upper()}")
    print(f"{'='*80}")
    print("Shape:", df.shape)
    print("\nColumns:")
    print(df.columns.tolist())
    print("\nDtypes:")
    print(df.dtypes)
    print("\nMissing values:")
    print(df.isnull().sum())
    print("\nDuplicate rows:", df.duplicated().sum())
    print("\nHead:")
    display(df.head())
```

```python
dfs = {
    "programs": programs,
    "users": users,
    "applications": applications,
    "pods": pods,
    "project_groups": project_groups,
    "ta_assignments": ta_assignments,
    "daily_engagement": daily_engagement,
    "assessments": assessments,
    "surveys": surveys,
    "dropouts": dropouts,
    "support_requests": support_requests,
    "ta_feedback": ta_feedback,
    "pod_checkins": pod_checkins
}

for name, df in dfs.items():
    audit_df(df, name)
```

## Cleaning Metadata

```python
programs.head()
```

```python
programs = programs.drop_duplicates()

date_cols = ["start_date", "end_date"]
for col in date_cols:
    programs[col] = pd.to_datetime(programs[col], errors="coerce")

programs["year"] = programs["year"].astype("Int64")
programs["num_students"] = programs["num_students"].astype("Int64")
programs["num_tas"] = programs["num_tas"].astype("Int64")

programs["delivery_mode"] = programs["delivery_mode"].str.strip().str.lower()
programs["region_focus"] = programs["region_focus"].str.strip().str.lower()

programs.head()
```

```python
users = users.drop_duplicates()

users["role"] = users["role"].str.strip().str.lower()
users["country"] = users["country"].str.strip()
users["timezone"] = users["timezone"].str.strip()
users["gender"] = users["gender"].fillna("Prefer not to say").str.strip()
users["education_level"] = users["education_level"].str.strip()
users["field_of_study"] = users["field_of_study"].str.strip()
users["preferred_language"] = users["preferred_language"].str.strip()
users["interest_area"] = users["interest_area"].str.strip()
users["availability_slot"] = users["availability_slot"].str.strip()

users["years_experience"] = pd.to_numeric(users["years_experience"], errors="coerce").fillna(0).astype(int)
users["prior_neuromatch"] = pd.to_numeric(users["prior_neuromatch"], errors="coerce").fillna(0).astype(int)

# Valid role check
valid_roles = ["student", "ta", "instructor"]
users = users[users["role"].isin(valid_roles)]

users.head()
```

```python
applications = applications.drop_duplicates()

applications["application_date"] = pd.to_datetime(applications["application_date"], errors="coerce")

score_cols = ["motivation_score", "cv_score", "technical_score", "diversity_score"]
for col in score_cols:
    applications[col] = pd.to_numeric(applications[col], errors="coerce").clip(0, 100)

binary_cols = ["selected", "waitlisted"]
for col in binary_cols:
    applications[col] = pd.to_numeric(applications[col], errors="coerce").fillna(0).astype(int)

applications["final_status"] = applications["final_status"].str.strip().str.lower()
applications["preferred_program_track"] = applications["preferred_program_track"].astype(str).str.strip()
applications["preferred_timeslot"] = applications["preferred_timeslot"].astype(str).str.strip()

applications.head()
```

```python
pods = pods.drop_duplicates()

pods["year"] = pd.to_numeric(pods["year"], errors="coerce").astype("Int64")
pods["num_students"] = pd.to_numeric(pods["num_students"], errors="coerce").astype("Int64")
pods["region_mix_score"] = pd.to_numeric(pods["region_mix_score"], errors="coerce").clip(0, 1)

pods["timeslot"] = pods["timeslot"].astype(str).str.strip()
pods["language"] = pods["language"].astype(str).str.strip()
pods["interest_cluster"] = pods["interest_cluster"].astype(str).str.strip()

pods.head()
```

```python
project_groups = project_groups.drop_duplicates()

project_groups["num_students"] = pd.to_numeric(project_groups["num_students"], errors="coerce").astype("Int64")
project_groups["group_topic"] = project_groups["group_topic"].astype(str).str.strip()

project_groups.head()
```

```python
ta_assignments = ta_assignments.drop_duplicates()

ta_assignments["ta_role"] = ta_assignments["ta_role"].astype(str).str.strip().str.lower()

valid_ta_roles = ["regular_ta", "project_ta"]
ta_assignments = ta_assignments[ta_assignments["ta_role"].isin(valid_ta_roles)]

ta_assignments.head()
```

```python
daily_engagement = daily_engagement.drop_duplicates()

daily_engagement["day"] = pd.to_numeric(daily_engagement["day"], errors="coerce").astype("Int64")

binary_cols = ["attended_curriculum", "attended_project", "assignments_completed"]
for col in binary_cols:
    daily_engagement[col] = pd.to_numeric(daily_engagement[col], errors="coerce").fillna(0).astype(int)

num_cols = ["curriculum_hours_attended", "project_hours_attended", "zoom_minutes", "forum_messages", "help_requests"]
for col in num_cols:
    daily_engagement[col] = pd.to_numeric(daily_engagement[col], errors="coerce").fillna(0)

# clip realistic bounds
daily_engagement["curriculum_hours_attended"] = daily_engagement["curriculum_hours_attended"].clip(0, 4.5)
daily_engagement["project_hours_attended"] = daily_engagement["project_hours_attended"].clip(0, 3.0)
daily_engagement["zoom_minutes"] = daily_engagement["zoom_minutes"].clip(0, 480)

daily_engagement.head()
```

```python
assessments = assessments.drop_duplicates()

assessments["score"] = pd.to_numeric(assessments["score"], errors="coerce").clip(0, 100)
assessments["date"] = pd.to_datetime(assessments["date"], errors="coerce")
assessments["type"] = assessments["type"].astype(str).str.strip().str.lower()

valid_assessment_types = ["pre", "mid", "post"]
assessments = assessments[assessments["type"].isin(valid_assessment_types)]

assessments.head()
```

```python
surveys = surveys.drop_duplicates()

surveys["survey_type"] = surveys["survey_type"].astype(str).str.strip().str.lower()
surveys["response_date"] = pd.to_datetime(surveys["response_date"], errors="coerce")
surveys["open_feedback"] = surveys["open_feedback"].fillna("").astype(str)

rating_cols = ["satisfaction", "difficulty", "content_quality", "ta_support"]
for col in rating_cols:
    surveys[col] = pd.to_numeric(surveys[col], errors="coerce").clip(1, 5)

valid_survey_types = ["pre", "mid", "post"]
surveys = surveys[surveys["survey_type"].isin(valid_survey_types)]

surveys.head()
```

```python
dropouts = dropouts.drop_duplicates()

dropouts["dropout_day"] = pd.to_numeric(dropouts["dropout_day"], errors="coerce").astype("Int64")
dropouts["reason"] = dropouts["reason"].astype(str).str.strip().str.lower()

valid_dropout_reasons = ["low_engagement", "time_constraint", "difficulty"]
dropouts = dropouts[dropouts["reason"].isin(valid_dropout_reasons)]

dropouts.head()
```

```python
support_requests = support_requests.drop_duplicates()

support_requests["channel"] = support_requests["channel"].astype(str).str.strip().str.lower()
support_requests["issue_type"] = support_requests["issue_type"].astype(str).str.strip().str.lower()

support_requests["response_time_hours"] = pd.to_numeric(
    support_requests["response_time_hours"], errors="coerce"
).fillna(0).clip(0, 168)

support_requests["resolved"] = pd.to_numeric(
    support_requests["resolved"], errors="coerce"
).fillna(0).astype(int)

support_requests["satisfaction"] = pd.to_numeric(
    support_requests["satisfaction"], errors="coerce"
).clip(1, 5)

support_requests.head()
```

```python
ta_feedback = ta_feedback.drop_duplicates()

ta_feedback["ta_role"] = ta_feedback["ta_role"].astype(str).str.strip().str.lower()

ta_feedback["avg_student_rating"] = pd.to_numeric(
    ta_feedback["avg_student_rating"], errors="coerce"
).clip(1, 5)

ta_feedback["num_sessions_taken"] = pd.to_numeric(
    ta_feedback["num_sessions_taken"], errors="coerce"
).fillna(0).astype(int)

ta_feedback["response_time_avg"] = pd.to_numeric(
    ta_feedback["response_time_avg"], errors="coerce"
).fillna(0)

ta_feedback["engagement_score"] = pd.to_numeric(
    ta_feedback["engagement_score"], errors="coerce"
).clip(0, 1)

ta_feedback.head()
```

```python
pod_checkins = pod_checkins.drop_duplicates()

pod_checkins["day"] = pd.to_numeric(pod_checkins["day"], errors="coerce").astype("Int64")
pod_checkins["completed"] = pd.to_numeric(pod_checkins["completed"], errors="coerce").fillna(0).astype(int)
pod_checkins["issues_reported"] = pd.to_numeric(pod_checkins["issues_reported"], errors="coerce").fillna(0).astype(int)
pod_checkins["mood_score"] = pd.to_numeric(pod_checkins["mood_score"], errors="coerce").clip(1, 5)

pod_checkins.head()
```

```python
programs.to_csv(f"{CLEAN_DIR}/programs_clean.csv", index=False)
users.to_csv(f"{CLEAN_DIR}/users_clean.csv", index=False)
applications.to_csv(f"{CLEAN_DIR}/applications_clean.csv", index=False)
pods.to_csv(f"{CLEAN_DIR}/pods_clean.csv", index=False)
project_groups.to_csv(f"{CLEAN_DIR}/project_groups_clean.csv", index=False)
ta_assignments.to_csv(f"{CLEAN_DIR}/ta_assignments_clean.csv", index=False)
daily_engagement.to_csv(f"{CLEAN_DIR}/daily_engagement_clean.csv", index=False)
assessments.to_csv(f"{CLEAN_DIR}/assessments_clean.csv", index=False)
surveys.to_csv(f"{CLEAN_DIR}/surveys_clean.csv", index=False)
dropouts.to_csv(f"{CLEAN_DIR}/dropouts_clean.csv", index=False)
support_requests.to_csv(f"{CLEAN_DIR}/support_requests_clean.csv", index=False)
ta_feedback.to_csv(f"{CLEAN_DIR}/ta_feedback_clean.csv", index=False)
pod_checkins.to_csv(f"{CLEAN_DIR}/pod_checkins_clean.csv", index=False)

print("✅ All cleaned CSVs saved to data/cleaned/")
```
